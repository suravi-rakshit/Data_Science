```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data=pd.read_excel("data science.xlsx")
print(data)
sns.scatterplot(data=data,x="longd",y="latd",label="city areas")
sns.scatterplot(data=data,x="longd",y="latd")
plt.xlabel("longd",fontsize=20)
plt.ylabel("latd",fontsize=20)
plt.title("AREA AND POPLUTAION OF CALIFORNIA CITIES",fontsize=20)
plt.legend(["200km","300km","500km"])
plt.show()
```
